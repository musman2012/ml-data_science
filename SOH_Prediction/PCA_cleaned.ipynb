## Using Principal Component Analysis (PCA) for Dimenstionality Reduction
Working with large dataset having multiple features affects the efficiency of prediction models in machine learning. Therefore, dimensionality reduction is mostly applied on the data to find independent features which can losslessly represent the data. PCA is one of such techniques which is mostly used to obtain those independent features. Use [this](https://medium.com/@aptrishu/understanding-principle-component-analysis-e32be0253ef0) link to get a better understanding about what PCA is and what are its benefits. Following code applies PCA technique for finding low variant features in the given batteries' dataset.

```python
'''
authored by Muhammad Usman (musman14@student.bradford.ac.uk)
'''
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

FILE_PATH = 'C:/Users/musman14/Desktop/DS Research/train_naive_encode_corr.csv'
bat_data = pd.read_csv(FILE_PATH)    # reading the data from csv file
```

Implementing PCA in Python primarily composed of three main functions:
1. Scaling the data (Features of different scale can be projected on the same scale)
2. Separate out target features
3. Use PCA to project the data in less dimenstions (1D in our case)
Details about each of these can be found [here](https://towardsdatascience.com/pca-using-python-scikit-learn-e653f8989e60).



```python
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# initializing standard scaler, to be used for data scaling
scaler = StandardScaler()          
```

```python
# dropping the probable target attributes from the data
output_data = bat_data.loc[:, ['max_soh_cell', 'min_soh', 'soh', 'max_soh', 'imbalance_percent']]
bat_data = bat_data.drop(columns = ['Unnamed: 0', 'max_soh_cell', 'min_soh', 'soh', 'max_soh', 'imbalance_percent'])

# backing up the data and scaling it to be used in PCA
bat_data_scaled = bat_data.copy()
bat_data_scaled[bat_data_scaled.columns] = scaler.fit_transform(bat_data_scaled)
```

```python
# selecing only one principal component from the analysis as we are already dealing with a group of three
pca = PCA(n_components=1)

# iteratubg the list of attributes to get combinations of three
attr_dict = dict()   # this dictionary will hold the variance of group of attributes in dict[attributes-name] = 1st PCA comp

# making group of three attributes
for i in range (len(bat_data_scaled.columns) - 3):
    for j in range (i+1, len(bat_data_scaled.columns) - 2):
        for k in range (j+1, len(bat_data_scaled.columns) - 1):
            input_cols = [bat_data_scaled.columns[i], bat_data_scaled.columns[j], bat_data_scaled.columns[k]]
            col_str = "" + bat_data_scaled.columns[i] + "," + bat_data_scaled.columns[j] + "," + bat_data_scaled.columns[k]
            pca_batt = pca.fit_transform(bat_data_scaled[input_cols]) # applying PCA
            var = np.round(pca.explained_variance_ratio_, decimals=3) * 100 # variance on the first component
            print(col_str,  "=" , var)
            attr_dict[col_str] = var
        
```

```python
import operator
# sorting the dictionary with respect to values on variance
sorted_dict = {k: v for k, v in sorted(attr_dict.items(), key=lambda item: item[1])}
single_attr_dict = dict()

count = 0

# going through top 500 results, with least variance
for key in sorted_dict:  
    temp = key.split(",")  # Splitting the keys in single attributes
    for val in temp:       # for each attribute, count how many times it is present in the top 500 results
        if val in single_attr_dict:
            single_attr_dict[val] += 1
        else:
            single_attr_dict[val] = 1
    if count == 500:
        break
    count+=1

```

```python
# plotting the results in bargraphs
df = pd.DataFrame([single_attr_dict])

plt.xticks(rotation='vertical')
plt.bar(single_attr_dict.keys(), single_attr_dict.values(), width=0.5, color='g')

#sns.countplot(x=df[:])
```
